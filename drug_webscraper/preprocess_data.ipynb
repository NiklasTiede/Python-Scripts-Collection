```python
import pandas as pd
```

```python
file = 'Compilation_of_CDER_NME_and_New_Biologic_Approvals_1985-2019.csv'
df = pd.read_csv(file)
df.head(4)
```

```python
# select only first two rows
preproc_df = df[['Proprietary  Name', 'Active Ingredient/Moiety']]
preproc_df.tail(20)
```

```python
# how many rows without name?
preproc_df.loc[df['Proprietary  Name'] == '[drug marketed without a proprietary name]']
```

```python
# these rows are cut out
preproc_df = preproc_df[preproc_df['Proprietary  Name'] != '[drug marketed without a proprietary name]']
```

```python
# check:
preproc_df.tail(30)
```

```python
# writing processed data to a file
preproc_df.to_csv('preprocessed_compilation.csv', index=False)
```

```python
file = 'fda_drugs_inchi.csv'
df = pd.read_csv(file)
df.head(4)
```

```python
df.columns
```

```python
len('InChI=1S/H2O4S/c1-5(2,3)4/h(H2,1,2,3,4)/p-2')
```

```python
# all inchis < 60 characters are filtered, to filter 'bad request' rows and salts
filter = df['InChI'].apply(len) > 60
final_df = df[filter]
```
